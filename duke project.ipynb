```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.bikewale.com/ktm-bikes/duke/"
```

```python
duke=requests.get(url)
```

```python
duke
```

```python
duke=BeautifulSoup(duke.text)
```

```python
duke
```

```python
model=duke.select('#bikeModeleList .model-card__label')
```

```python
model
```

```python
price=duke.select('#bikeModeleList .font-bold')
```

```python
price
```

```python
import pandas as ps
```

```python
dukelist=ps.DataFrame(model,columns=["modelname"])
dukelist["modelprice"]=price
```

```python
dukelist
```

```python
dukelist.to_csv('D:\duke.csv')
```
